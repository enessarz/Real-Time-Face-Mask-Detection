```python
import matplotlib.pyplot as plt
import xml.etree.ElementTree as ET
import os
import cv2

image_dir = '../images'
annot_dir = '../annotations'

example_file = os.listdir(annot_dir)[0]
file_id = example_file.split('.')[0]

img_path = os.path.join(image_dir, file_id + '.png')
xml_path = os.path.join(annot_dir, file_id + '.xml')

img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

tree = ET.parse(xml_path)
root = tree.getroot()

for obj in root.findall('object'):
    label = obj.find('name').text
    
    bndbox = obj.find('bndbox')
    xmin = int(bndbox.find('xmin').text)
    ymin = int(bndbox.find('ymin').text)
    xmax = int(bndbox.find('xmax').text)
    ymax = int(bndbox.find('ymax').text)
                
    if label == "with_mask":
        color = (0, 255, 0)
    elif label == "without_mask":
        color = (255, 0, 0)
    else:
        color(255, 255, 0)
        
    cv2.rectangle(img, (xmin, ymin), (xmax, ymax), color, 2)
    cv2.putText(img, label, (xmin, ymin - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
    
    plt.figure(figsize=(10, 10))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

## Cropping

```python
image_dir = '../images'
annot_dir = '../annotations'
output_dir = '../dataset_crop'

labels = ['with_mask', 'without_mask', 'mask_weared_incorrect']

for label in labels: 
    path = os.path.join(output_dir, label)
    os.makedirs(path, exist_ok=True)
    
xml_files = os.listdir(annot_dir)
count = 0

for xml_file in xml_files:
    tree = ET.parse(os.path.join(annot_dir, xml_file)) 
    root = tree.getroot()
    
    file_name = root.find('filename').text
    img_path = os.path.join(image_dir, file_name)
    
    img = cv2.imread(img_path)
    if img is None:
        print(f"Error: {file_name} Could not readed, passing.")
        continue
    
    for obj in root.findall('object'):
        label = obj.find('name').text
        
        bndbox = obj.find('bndbox')
        xmin = int(bndbox.find('xmin').text)
        ymin = int(bndbox.find('ymin').text)
        xmax = int(bndbox.find('xmax').text)
        ymax = int(bndbox.find('ymax').text)
        
        face_img = img[max(0, ymin):min(img.shape[0],ymax), max(0, xmin):min(img.shape[1], xmax)]

        if face_img.size > 0 and label in labels:
            save_path = os.path.join(output_dir, label, f"face_{count}.jpg")
            cv2.imwrite(save_path, face_img)
            count += 1

print(f"Process is done total {count} face is cropped and added to the '{output_dir}' folder.")
```
